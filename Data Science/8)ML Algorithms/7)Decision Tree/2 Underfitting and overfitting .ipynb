```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
def analyzer(max_depth):
    # Load the dataset
    df = pd.read_csv('Social_Network_Ads.csv')
    X = df.iloc[:, 2:4].values
    y = df.iloc[:, -1].values

    # Initialize and train the classifier
    clf = DecisionTreeClassifier(max_depth=max_depth)
    clf.fit(X, y)

    # Create the meshgrid
    a = np.arange(start=X[:, 0].min() - 1, stop=X[:, 0].max() + 1, step=0.1)
    b = np.arange(start=X[:, 1].min() - 1, stop=X[:, 1].max() + 1, step=100)
    XX, YY = np.meshgrid(a, b)

    # Predict labels for each point in the meshgrid
    input_array = np.array([XX.ravel(), YY.ravel()]).T
    labels = clf.predict(input_array)

    # Plot the decision boundary
    plt.contourf(XX, YY, labels.reshape(XX.shape), alpha=0.5)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolor='k')
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.title(f'Decision Tree Classifier (max_depth={max_depth})')
    plt.show()
```

```python
analyzer(max_depth=None) #overfitting
```

```python
analyzer(max_depth=1) #underfitting
```
